```python
import pandas as pd
```

```python
# Simple Data Frame
data = {
    "Name":["alice","bob","charlie","david"],
    "Age":[25,30,22,35],
    "City":["New York","Paris","London","Berlian"],
    "Salary":[70000,80000,60000,90000]
}

df = pd.DataFrame(data)
print("Orginal DataFrame:\n",df)
```

```python
# Labled Based Selection
subset_loc = df.loc[1:3,["Name","Salary"]]
print("Subset with .loc:\n",subset_loc)
```

```python
subset_iloc = df.iloc[0:2,1:3]
print("Subset with .iloc:\n",subset_iloc)
```

```python
# boealen filtering 
high_salary = df[df["Salary"] > 75000]
print("High salary Filter:\n",high_salary)
```

```python
# safe assignment ;using .loc
df.loc[df["Age"] < 30,"Salary"] = 65000
print("Low Age Salary change :\n",df.loc[df["Age"]< 30, "Salary"])
```
